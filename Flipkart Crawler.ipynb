```python
#importing libraries

import os
import pandas as pd
import numpy as np 
import datetime
import time
import dateutil
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup as bs
import warnings
warnings.filterwarnings("ignore")
```

```python
#setting up web browser parameter
options = Options()
options.add_argument("--start-maximized")
#options.add_argument("--headless")
driver = webdriver.Chrome(options = options)


#crawler
driver.get("https://www.flipkart.com/")
time.sleep(3)  #page loading delay

cross_bar = driver.find_element_by_xpath('/html/body/div[2]/div/div/button')
cross_bar.click()
time.sleep(5)

search_bar = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_bar.click()
search_bar.send_keys('Samsung phones')
time.sleep(10)

#Search
search = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search.click()
```

```python
data = driver.page_source
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(data,'html.parser')
product_name = soup.findAll('div', {'class' : '_3wU53n'})
product_name = [x.text for x in product_name]
product_price = soup.findAll('div', {'class' : '_1vC4OE _2rQ-NK'})
product_price = [x.text for x in product_price]
data_list = [product_name,product_price]
```

```python
data_list = [product_name,product_price]
scraped_data = pd.DataFrame(data= data_list )
scraped_data = scraped_data.transpose()
scraped_data.columns = ['Product Name', 'Product Price']
```
